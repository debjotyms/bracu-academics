```python
# Step 1: Import Lib
import sympy as sm

# Step 2: Make a Symbol
x = sm.symbols('x')

# Step 3: Write the function
f = 2*x**3 + x**2 + 3*x + 5

# Step 4: Differentiate
df = sm.diff(f,x)
df2 = sm.diff(f,x,2)

# Step 5: Printing
print(sm.latex(df))
print(df2)

# Step 6: Value Substituting
# Exact SymPy expressions can be converted to floating-point approximations (decimal numbers) using either the .evalf() method or the N() function.
print(sm.N(f.subs(x,4)))
print(df.subs(x,4).evalf())
print(df2.subs(x,4).evalf())
```

```python
# Lambdify Function
# if eval() function is not efficient
# Translates Sympy expression into numerical expression
# Create options for evalutaing expression

f = sm.lambdify(x,f)
df = sm.lambdify(x,df)

print(f(4))
print(df(4))
```

```python
# Evaluate f in the interval [-5,6]

import numpy as np

test_numbers = np.arange(-5,6)
print(test_numbers)

a = df(test_numbers)
print(a)
```

```python
import matplotlib.pyplot as plt

plt.title('Z')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.plot(test_numbers, a)
```
